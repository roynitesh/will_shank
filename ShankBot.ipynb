```python
# Shankbot
def shankbot(n):
    i = 0
    x = 10
    while x%n != 1:
        i = i+1
        x = (x%n)*10
    return i+1
```

```python
# test if a number is prime
def prime_check(n):
    for i in range(2,n):
        if n%i==0:
            return "composite number"
            break
        else:
            pass
```

```python
# generate a list of prime numbers smaller than m
def prime_list(m):
    a = []
    for i in range(2,m):
        if prime_check(i) != 'composite number':
            a.append(i)
    a.remove(2)
    a.remove(5)
    return a
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(prime_list(2000),[shankbot(i) for i in prime_list(2000)])
```
